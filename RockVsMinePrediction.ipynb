```python
import kagglehub
import pandas as pd
import os

# Download latest version
dataset_dir = kagglehub.dataset_download("mayurdalvi/sonar-mine-dataset")

# Find the CSV file within the downloaded directory
for filename in os.listdir(dataset_dir):
    if filename.endswith(".csv"):
        csv_file_path = os.path.join(dataset_dir, filename)
        break  # Stop after finding the first CSV file

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(csv_file_path)
print("csv file path :",csv_file_path)

print("Path to dataset files:", dataset_dir)
print(df.head()) # Print first 5 rows of the dataframe to verify
```

Training model using logisticRegression

```python
x = df.drop("label",axis=1)
y = df['label']
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1,stratify=y,random_state=1)
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

#Model Evalution

Accuracy on Training Data

```python
x_train_pred = model.predict(x_train)
acc =  accuracy_score(x_train_pred,y_train)
print("Accuracy on training data :",acc)
```

Accuracy on test data

```python
x_test_pred = model.predict(x_test)
acc =  accuracy_score(x_test_pred,y_test)
print("Accuracy on testing data :",acc)
```
